```python
#Taking data for test purpose

df = results[(results['Team1'] == ' India ') | (results['Team2'] == ' India ')]
india = df.iloc[:]
india.head()
```

```python
#Filtering T20 records for only 12 teams {assumption: Ireland & Netherlands because they performed well in qualifiers 
#till record date}

worldcup_teams = [' England ', ' South Africa ', ' Ireland ', ' Zimbabwae ', 
            ' Pakistan ', ' New Zealand ', ' Sri Lanka ', ' Afghanistan ', 
            ' Australia ', ' Bangladesh ', ' India ',' Netherlands ']
df_teams_1 = results[results['Team1'].isin(worldcup_teams)]
df_teams_2 = results[results['Team2'].isin(worldcup_teams)]
df_teams = pd.concat((df_teams_1, df_teams_2))
df_teams.drop_duplicates()
df_teams.count()
```
