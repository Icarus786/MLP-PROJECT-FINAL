```python
#importing libraries

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import matplotlib.ticker as ticker
import matplotlib.ticker as plticker
from sklearn.model_selection import train_test_split
from sklearn.tree import DecisionTreeClassifier
from sklearn.datasets import make_classification
```

```python
#load data 

world_cup = pd.read_csv('T20Fixture.csv')
results = pd.read_csv('T20Records.csv')
ranks = pd.read_csv('T20Ranking.csv')
```

```python
world_cup.head()
```

```python
# Previous records
results
```

```python
#Removing the record with ties
data=results[results.Winner!='tie']
```

```python
# Dropping Venue column ,Since Only Australia will host the World Cup so home team advantage not possible for other team
data.drop(columns=['Venue'],inplace=True)
```

```python
#Dropping the date column
data.drop(columns=['Date'],inplace=True)
data.head()
```

```python
#Filtering the records for only teams that will play the world cup
worldcup_teams = [' England ', ' South Africa ', ' Ireland ', ' Zimbabwae ', 
            ' Pakistan ', ' New Zealand ', ' Sri Lanka ', ' Afghanistan ', 
            ' Australia ', ' Bangladesh ', ' India ',' Netherlands ']
df_teams_1 = data[data['Team1'].isin(worldcup_teams)]
df_teams_2 = data[data['Team2'].isin(worldcup_teams)]
df_teams = pd.concat((df_teams_1, df_teams_2))
df_teams.drop_duplicates()
df_teams.count()
```

```python
#Adding column 'Count': It will record the team which won the most matches from its previous 6 matches (i.e. if India have won 4 matches from its previous 6 matches and Pakistan have won 3 matches than count which select India)
# and 'Rank': The team which is having higher rank will be named in this column
df_teams['Count']=0
df_teams['Rank']=0
df_team= df_teams.reset_index(drop=True) 
df_team
```

```python
#filling the count and rank column
for i in range(700):
    dt1=df_team['Team1'].iloc[i]
    dt2=df_team['Team2'].iloc[i]
    c1=0
    c2=0
    w1=0
    w2=0
    for j in range(i+1,i+50):
        if ((df_team['Team1'].iloc[j]==dt1) or (df_team['Team2'].iloc[j]==dt1.rstrip())) and c1<6:
            if df_team['Winner'].iloc[j]==dt1.strip():
                w1=w1+1
                c1=c1+1
            else:
                c1=c1+1
    
    for j in range(i+1,i+50):
        if ((df_team['Team1'].iloc[j]==(dt2+ ' ')) or (df_team['Team2'].iloc[j]==dt2)) and c2<6:
            if df_team['Winner'].iloc[j]==dt2.strip():
                w2=w2+1
                c2=c2+1
            else:
                c2=c2+1
    if(w2>w1):
        df_team['Count'].iloc[i]=dt2
    else:
        df_team['Count'].iloc[i]=dt1
    r1=0
    r2=0
    for k in range(12):
        if (dt1.strip()==ranks['Team'].iloc[k]):
            r1=ranks['Rank'].iloc[k]
    for k in range(12):
        if (dt2.strip()==ranks['Team'].iloc[k]):
            r2=ranks['Rank'].iloc[k]
    if(r2>r1):
        df_team['Rank'].iloc[i]=dt1
    else:
        df_team['Rank'].iloc[i]=dt2
            
    
        
```

```python
data=df_team.iloc[0:700]
data
```

```python
#Assigning the binary number to data , team1: 0 and team2: 1
for i in range(700):
    dt1=data['Team1'].iloc[i]
    dt2=data['Team2'].iloc[i]
    data['Team1'].iloc[i]=0
    data['Team2'].iloc[i]=1
    if data['Winner'].iloc[i]==dt1.strip():
        data['Winner'].iloc[i]=data['Team1'].iloc[i]
    else:
        data['Winner'].iloc[i]=data['Team2'].iloc[i]
    
    if data['Count'].iloc[i]==dt1:
        data['Count'].iloc[i]=data['Team1'].iloc[i]
    else:
        data['Count'].iloc[i]=data['Team2'].iloc[i]
    
    if data['Rank'].iloc[i]==dt1:
        data['Rank'].iloc[i]=data['Team1'].iloc[i]
    else:
        data['Rank'].iloc[i]=data['Team2'].iloc[i]
        
    
        
```

```python
#Seperating the output i.e. Winner column for testing model
y=data["Winner"]
y = y.astype(float, errors = 'raise')
```

```python
#dropping the target column and creating features file
X=data.drop('Winner',axis=1)
X['Team1'] = X.Team1.astype(float)
X['Team2'] = X.Team2.astype(float)
X['Count'] = X.Count.astype(float)
X['Rank'] = X.Rank.astype(float)
```

```python
#importing libraries for SVM(Support Vector Machine) model
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.svm import SVC
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size= 0.2)
```

```python
model = SVC(kernel = 'linear', C = 1)
```

```python
model.fit(X_train, y_train)
```

```python
svm_pred = model.predict(X_test)
```

```python
#Accuracy score
accuracy = model.score(X_test, y_test)
accuracy
```

```python
# Create a dictionary with the points table for Group 1
group1 = {
    'Team': ['England', 'New Zealand'],
    'Win': [4, 4]
}

# Create a dictionary with the points table for Group 2
group2 = {
    'Team': ['India', 'Pakistan'],
    'Win': [5, 4]
}

# Create dataframes from the dictionaries
group1_df = pd.DataFrame(group1)
group2_df = pd.DataFrame(group2)

# Set the winners and runner-ups for Group 1
group1_winner = 'New Zealand'
group1_runnerup = 'England'

# Create X_train and y_train
X_train = np.array([
    group1_df.loc[group1_df['Team'] == group1_winner]['Win'],
    group2_df.loc[group2_df['Team'] == 'India']['Win']
])
X_train = X_train.reshape(2, 1)
y_train = np.array([1, 0])

# Train the decision tree classifier
dtc = DecisionTreeClassifier()
dtc.fit(X_train, y_train)

# Predict the winners of the semi-finals and final
semi_finalist_1 = group2_df.loc[group2_df['Team'] == 'Pakistan']['Team'].values[0]
semi_finalist_2 = group1_df.loc[group1_df['Team'] == group1_runnerup]['Team'].values[0]

X_test = np.array([
    group2_df.loc[group2_df['Team'] == semi_finalist_1]['Win'],
    group1_df.loc[group1_df['Team'] == group1_winner]['Win']
])
X_test = X_test.reshape(2, 1)
finalist_1 = dtc.predict(X_test)[0]
finalist_1 = semi_finalist_1 if finalist_1 == 1 else group1_winner
print('Winner of first semi-final:', finalist_1)

X_test = np.array([
    group2_df.loc[group2_df['Team'] == 'India']['Win'],
    group1_df.loc[group1_df['Team'] == semi_finalist_2]['Win']
])
X_test = X_test.reshape(2, 1)
finalist_2 = dtc.predict(X_test)[0]
finalist_2 = 'India' if finalist_2 == 1 else semi_finalist_2
print('Winner of second semi-final:', finalist_2)

X_test = np.array([
    group2_df.loc[group2_df['Team'] == finalist_1]['Win'],
    group1_df.loc[group1_df['Team'] == finalist_2]['Win']
])
X_test = X_test.reshape(2, 1)
winner = dtc.predict(X_test)[0]
winner = finalist_1 if winner == 1 else finalist_2

# Print the winner of the tournament
print('The winner of the ICC T20 World Cup 2022 is:', winner)
```
